```python
# FIRST LET'S CREATE A RANDOM DATAFRAME

import random
import pandas as pd
eleves=['William','Hugh','Marvin','Stella','Serge','Kimiko']
# LIST COMPREHENSION
liste_eleves = [random.choice(eleves) for _ in range(60)]

# as we saw before, this is EQUIVALENT TO
# liste_eleves=[]
# for _ in range(60):
#    liste_eleves.append(random.choice(eleves))

def note():
    return(random.randint(1,20))

notes = [ note() for _ in range(60)]

eleves_notes = pd.DataFrame()
eleves_notes['eleve']=liste_eleves
eleves_notes['note']=notes
eleves_notes
```

```python

# Like we did last TP, let's create a function to extract the best 5 grades for each student

def get_top5(df):
    return df.sort_values(by="note")[-5:]

# rank here is static, it cannot be transmitted during the function call

eleves_notes.groupby(['eleve']).apply(get_top5)
```

```python
# During last TP, a student proposed this method :
# **kwags will create a dict with all the named parameters

def get_top(df,**kwags):
    return df.sort_values(by="note")[-kwags['top']:]

eleves_notes.groupby(['eleve']).apply(get_top,top=5)
```

```python
# please note that kwags name is not mandatory
# the only mandatory element is the double * (splats) for "packing"
# Here we add the sort column name as a parameter.

def get_top(df,**mondicoamoi):  
    return df.sort_values(by=mondicoamoi['sort'])[-mondicoamoi['top']:]

eleves_notes.groupby(['eleve']).apply(get_top,sort="note",top=3)
```

```python
# But we can do more simple ( and add sort column as parameter) :

def get_top(df,top,sort):    
    return df.sort_values(by=sort)[-top:]

eleves_notes.groupby(['eleve']).apply(get_top,sort="note",top=2)
```

```python
# And give default values

def get_top(df,top=10,sort='note'):    
    return df.sort_values(by=sort)[-top:]

eleves_notes.groupby(['eleve']).apply(get_top,top=2)
```

# Display a choropleth map with number of girls named Emma born in 2018

```python
df_Emma_2018=pd.DataFrame()
df_Emma_2018["nb_Emma"]=prenoms[(prenoms.preusuel=="EMMA") & (prenoms.annais=="2018")].groupby("dpt").nombre.sum()
df_Emma_2018
```

```python
df_Emma_2018=df_Emma_2018.reset_index()
df_Emma_2018
#On remarque qu'il manque un département surement il n'ya vait pas d'EMMA dans ce dpt en 2018
```

```python
df_Emma_2018["code_insee"]=df_Emma_2018["dpt"]
map_df=map_df[map_df.code_insee.str.contains("\d\d\d")==False]
merged = map_df.merge(df_Emma_2018, on='code_insee')
merged.head()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, figsize=(8, 6))

merged.plot(column="nb_Emma",   ax=ax, linewidth=0.3,edgecolor="0.8",legend=True,legend_kwds
            ={"label": "Number of girls named Emma born in 2018"})
#Région ou il n'y a pas d'EMMA représenté en blanc
```

# The same but weighted

```python
df_total_by_dpt=pd.DataFrame()
df_total_by_dpt["nb_total_2018_by_dpt"]=prenoms[prenoms.annais=="2018"].groupby("dpt").nombre.sum()
df_total_by_dpt=df_total_by_dpt.reset_index()
df_total_by_dpt
```

```python
df_Emma_2018_weighted=df_Emma_2018.merge(df_total_by_dpt,on='dpt')
df_Emma_2018_weighted["nb_Emma_weighted"]=df_Emma_2018_weighted["nb_Emma"]/df_Emma_2018_weighted["nb_total_2018_by_dpt"]
df_Emma_2018_weighted
```

```python
merged2 = map_df.merge(df_Emma_2018_weighted, on='code_insee')
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, figsize=(8, 6))

merged2.plot(column="nb_Emma_weighted",   ax=ax, linewidth=0.3,edgecolor="0.8",legend=True,legend_kwds
            ={"label": "Number of girls named Emma born in 2018 weighted"})
#On voit une amélioration de l'analyse qui aurait pu être mal interprété au premier graphe
```
