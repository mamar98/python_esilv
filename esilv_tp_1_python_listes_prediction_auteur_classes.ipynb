# 1.6) Durée d'insertion en queue / tête de liste

## 1.6.1) fonction de mesure de temps de calcul

```python
import time
import pandas as pd
```

```python
def duree(debut):
    return time.time() - debut
```

## 1.6.2) temps d'insertion en queue de liste

```python

conclusion = u"L'insertion en queue de liste est en temps constant"
liste      = list()
durees     = {}
for i in range(100000):
    debut = time.time()
    liste.append(i)
    fin   = time.time()
    durees[i] = fin - debut
    
df = pd.DataFrame.from_dict(durees, orient='index')    
df.plot(title=conclusion)
```

## 1.6.3) temps d'insertion en tête de liste

```python
conclusion         = u"L'insertion en tête de liste est en temps linéaire"
liste              = list()
position_insertion = 0
durees             = {}
for i in range(100000):
    debut = time.time()
    liste.insert(position_insertion, i)
    fin = time.time()
    durees[i] = fin - debut
    
df = pd.DataFrame.from_dict(durees, orient='index')    
df.plot(title=conclusion)
```

## 1.6.4) La classe "dequeue"

```python
from collections import deque
conclusion         = u"""
Deque insert en tête de liste en temps constant
et en étant  un ordre de magnitude plus rapide.
"""
double_queue       = deque()
position_insertion = 0
durees             = {}
for i in range(100000):
    debut = time.time()
    double_queue.appendleft(i)
    fin = time.time()
    durees[i] = fin - debut
    
df = pd.DataFrame.from_dict(durees, orient='index')    
df.plot(title=conclusion)
```

```python
from collections import deque
conclusion         = u"""
Deque insert en queue de liste en temps constant
et en étant  un ordre de magnitude plus rapide.
"""
double_queue       = deque()
position_insertion = 0
durees             = {}
for i in range(100000):
    debut = time.time()
    double_queue.append(i)
    fin = time.time()
    durees[i] = fin - debut
    
df = pd.DataFrame.from_dict(durees, orient='index')    
df.plot(title=conclusion)
```

## Calcul d'une variation dans une liste de chiffres

```python
r = []
for position, valeur_today in enumerate(cours_de_bourse): # enumerate renvoie 2 valeurs, stockées dans position et valeur   
    if position > 0:
        valeur_veille = cours_de_bourse[position - 1]
        variation     = (valeur_today - valeur_veille )/valeur_veille *100
        r.append(variation)
pd.DataFrame(r).plot()
```

## 7.2) concaténer les textes pour chaque auteur

```python
get_word = nltk.corpus.gutenberg.words 
```

```python
get_word() # fonction pour récupérer les mots pour un texte donné
```

```python
def get_top_n(liste_de_mots, taille_minimale, n):
    from collections import defaultdict
    dico = defaultdict(int)
    for mot in liste_de_mots:
        dico[mot]+=1
    
    mots_dans_l_ordre_d_usage      = sorted(dico, key= lambda x: dico[x], reverse=True)

    mot_tries_de_plus_de_n_lettres = filter(lambda mot: len(mot)>taille_minimale, 
                                             mots_dans_l_ordre_d_usage)
    
    mot_tries_de_plus_de_n_lettres = list(mot_tries_de_plus_de_n_lettres) 
    
    
    r = mot_tries_de_plus_de_n_lettres
    return r[:n]
```

```python
dico_auteurs.items()
```

```python
auteur_Vs_texte = {}
for auteur, liste_de_textes in dico_auteurs.items():
    mots = []
    for nom_fichier in liste_de_textes :
        mots.extend(get_word(nom_fichier))
        
    auteur_Vs_texte[auteur] = get_top_n(mots, 5, 100)
```

```python
database = pd.DataFrame(auteur_Vs_texte)
```

```python
database
```

# 8) classes

```python
class oeuvre():
    def __init__(self, auteur, titre ):
        """
        Méthode d'instanciation
        """
        self.auteur = auteur
        self.titre  = titre
    def __str__(self):
        """
        Méthode d'affichage
        """
        return f"{self.auteur} : {self.titre}"
```

```python

guerre_et_paix = oeuvre("Tolstoi", "guerre et paix")
print(guerre_et_paix)
```

# 8.1) héritage de classes

```python
class sculpture(oeuvre): # sculpture hérite de la classe oeuvre
    def __init__(self, auteur, titre, matériel):
        self.matériel = matériel
        super(sculpture, self).__init__(auteur, titre) # Appel de la fonction d'initialisation de la classe mère
        
    def __str__(self): # Surcharge de la méthode print
        return super(sculpture, self).__str__() + \
    " sculpture en  "+ self.matériel

```

```python
class livre(oeuvre): # livre hérite de la classe oeuvre
    
    def __init__(self, auteur, titre, nb_de_pages):
        self.nb_de_pages = nb_de_pages
        super(livre, self).__init__(auteur, titre) # Appel de la fonction d'initialisation de la classe mère
    
    def __str__(self): # Surcharge de la méthode print
        return f"""
        {self.titre} de {self.auteur} avec {self.nb_de_pages} pages
        """
```

```python

le_penseur = sculpture("Rodin", "le penseur", "bronze")
print(le_penseur)
```

```python
guerre_et_paix = livre("Tolstoi", "guerre et paix", 1000)
```
